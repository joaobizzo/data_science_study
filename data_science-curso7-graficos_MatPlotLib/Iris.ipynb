```python
import numpy as pd
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('iris.csv')
```

```python
df.head()
```

```python
fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0,0,1,1])

eixo.scatter(df['comprimento_sépala'], df['largura_sépala'])
eixo.set_title('Gráfico de Dispersão', fontsize=20)
eixo.set_xlabel('Comprimento da Sépala', fontsize=16)
eixo.set_ylabel('Largura da Sépala', fontsize=16)
```

```python
fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0,0,1,1])

eixo.scatter(df['comprimento_sépala'], df['largura_sépala'])
eixo.set_title('Gráfico de Dispersão', fontsize=25)
eixo.set_xlabel('Comprimento da Sépala', fontsize=20)
eixo.set_ylabel('Largura da Sépala', fontsize=20)
eixo.tick_params(labelsize=15)
```

```python
fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0,0,1,1])

cores = {'Iris-setosa': 'r', 'Iris-versicolor': 'b', 'Iris-virginica': 'g'}

for especie in df['espécie'].unique():
    tmp = df[df['espécie'] == especie]
    eixo.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'],
                color=cores[especie])

eixo.set_title('Gráfico de Dispersão', fontsize=25)
eixo.set_xlabel('Comprimento da Sépala', fontsize=20)
eixo.set_ylabel('Largura da Sépala', fontsize=20)
eixo.tick_params(labelsize=15)
eixo.legend(cores, fontsize=20)
```

```python
fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0,0,1,1])

cores = {'Iris-setosa': 'r', 'Iris-versicolor': 'b', 'Iris-virginica': 'g'}
marcadores = {'Iris-setosa': 'x', 'Iris-versicolor': 'o', 'Iris-virginica': 'v'}

for especie in df['espécie'].unique():
    tmp = df[df['espécie'] == especie]
    eixo.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'],
                color=cores[especie], marker=marcadores[especie])

eixo.set_title('Gráfico de Dispersão', fontsize=25)
eixo.set_xlabel('Comprimento da Sépala', fontsize=20)
eixo.set_ylabel('Largura da Sépala', fontsize=20)
eixo.tick_params(labelsize=15)
eixo.legend(cores, fontsize=20)
```

```python
fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0,0,1,1])

cores = {'Iris-setosa': 'r', 'Iris-versicolor': 'b', 'Iris-virginica': 'g'}
marcadores = {'Iris-setosa': 'x', 'Iris-versicolor': 'o', 'Iris-virginica': 'v'}

for especie in df['espécie'].unique():
    tmp = df[df['espécie'] == especie]
    eixo.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'],
                color=cores[especie], marker=marcadores[especie],
                s=100)

eixo.set_title('Gráfico de Dispersão', fontsize=25)
eixo.set_xlabel('Comprimento da Sépala', fontsize=20)
eixo.set_ylabel('Largura da Sépala', fontsize=20)
eixo.tick_params(labelsize=15)
eixo.legend(cores, fontsize=20)
```

```python
fig = plt.figure(figsize=(8, 5))
eixo = fig.add_axes([0,0,1,1])

eixo.boxplot(df.drop('espécie', axis=1).values, patch_artist=True)
eixo.set_title('Gráfico de Caixa', fontsize=15)
eixo.set_xticklabels(df.drop('espécie', axis=1).columns, fontsize=10)
```

```python
fig = plt.figure(figsize=(8, 5))
eixo = fig.add_axes([0,0,1,1])

cores = ['red', 'orange', 'yellow', 'green']

caixas = eixo.boxplot(df.drop('espécie', axis=1).values, patch_artist=True)
eixo.set_title('Gráfico de Caixa', fontsize=15)
eixo.set_xticklabels(df.drop('espécie', axis=1).columns, fontsize=10)

for caixa, cor in zip(caixas['boxes'], cores):
    caixa.set(color=cor)
```

```python
fig = plt.figure(figsize=(8, 5))
eixo = fig.add_axes([0,0,1,1])

cores = ['indigo', 'darkblue', 'blue', 'darkgreen']

caixas = eixo.boxplot(df.drop('espécie', axis=1).values, patch_artist=True)
eixo.set_title('Gráfico de Caixa', fontsize=15)
eixo.set_xticklabels(df.drop('espécie', axis=1).columns, fontsize=10)

for caixa, cor in zip(caixas['boxes'], cores):
    caixa.set(color=cor)
    
for outlier in caixas['fliers']:
    outlier.set(marker='x', markersize=8)
```

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0,0,1,1])

eixo.hist(df['comprimento_pétala'])
eixo.set_title('Histograma Comprimento da Pétala', fontsize=15)
eixo.set_xlabel('Comprimento da Pétala', fontsize=15)
eixo.grid(True)
```

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0,0,1,1])

eixo.hist(df['comprimento_pétala'], bins=20)
eixo.set_title('Histograma Comprimento da Pétala', fontsize=15)
eixo.set_xlabel('Comprimento da Pétala', fontsize=15)
eixo.grid(True)
```

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0,0,1,1])

eixo.hist(df['comprimento_pétala'], bins=20, density=True)
eixo.set_title('Histograma Comprimento da Pétala', fontsize=15)
eixo.set_xlabel('Comprimento da Pétala', fontsize=15)
eixo.grid(True)
```

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0,0,1,1])

mu, sigma = df['comprimento_pétala'].mean(),df['comprimento_pétala'].std()

eixo.hist(df['comprimento_pétala'], bins=20)
eixo.set_title('Histograma Comprimento da Pétala', fontsize=15)
eixo.set_xlabel('Comprimento da Pétala', fontsize=15)
eixo.grid(True)

eixo.annotate(f'$mu = {mu:.2f}$\n$sigma = {sigma:.2f}$',
               xy=(4.5, 20), fontsize=20)
```

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0,0,1,1])

mu, sigma = df['comprimento_pétala'].mean(),df['comprimento_pétala'].std()

eixo.hist(df['comprimento_pétala'], bins=20)
eixo.set_title('Histograma Comprimento da Pétala', fontsize=15)
eixo.set_xlabel('Comprimento da Pétala', fontsize=15)
eixo.grid(True)

eixo.annotate(f'$mu = {mu:.2f}$\n$sigma = {sigma:.2f}$',
               xy=(4.5, 20), fontsize=20)

eixo.axvline(mu, color='k', linestyle='--')
eixo.annotate('média', xy=(mu-1.3, 28), fontsize=20)
```

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0,0,1,1])

mu, sigma = df['comprimento_pétala'].mean(),df['comprimento_pétala'].std()

eixo.hist(df['comprimento_pétala'], bins=20)
eixo.set_title('Histograma Comprimento da Pétala', fontsize=15)
eixo.set_xlabel('Comprimento da Pétala', fontsize=15)
eixo.grid(True)

eixo.annotate(f'$\mu = {mu:.2f}$\n$\sigma = {sigma:.2f}$',
               xy=(4.5, 20), fontsize=20)

eixo.axvline(mu, color='k', linestyle='--')
eixo.annotate('média', xy=(mu-1.3, 28), fontsize=15)

eixo.axvline(df['comprimento_pétala'].median(), color='g', linestyle='--')
eixo.annotate('mediana', xy=(df['comprimento_pétala'].median(), 31), fontsize=15)
```

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0,0,1,1])

df_iv = df[df['espécie'] == 'Iris-versicolor']

mu, sigma = df_iv['comprimento_pétala'].mean(),df_iv['comprimento_pétala'].std()

eixo.hist(df_iv['comprimento_pétala'], bins=20)
eixo.set_title('Íris-Versicolor / Histograma Comprimento da Pétala', fontsize=15)
eixo.set_xlabel('Comprimento da Pétala', fontsize=15)
eixo.grid(True)

eixo.annotate(f'$\mu = {mu:.2f}$\n$\sigma = {sigma:.2f}$',
               xy=(4.6, 6), fontsize=20)

eixo.axvline(mu, color='k', linestyle='--')
eixo.annotate('média', xy=(mu-0.5, 5.5), fontsize=15)

eixo.axvline(df_iv['comprimento_pétala'].median(), color='r', linestyle='--')
eixo.annotate('mediana', xy=(df_iv['comprimento_pétala'].median()-0.7, 6.5), fontsize=15,
             color='r')

fig.savefig('histograma_iv.png', bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0,0,1,1])

df_is = df[df['espécie'] == 'Iris-setosa']

mu, sigma = df_is['comprimento_pétala'].mean(),df_is['comprimento_pétala'].std()

eixo.hist(df_is['comprimento_pétala'], bins=20)
eixo.set_title('Íris-Setosa / Histograma Comprimento da Pétala', fontsize=15)
eixo.set_xlabel('Comprimento da Pétala', fontsize=15)
eixo.grid(True)

eixo.annotate(f'$\mu = {mu:.2f}$\n$\sigma = {sigma:.2f}$',
               xy=(1.7, 7), fontsize=20)

eixo.axvline(mu, color='k', linestyle='--')
eixo.annotate('média', xy=(mu-0.2, 13.5), fontsize=15)

eixo.axvline(df_is['comprimento_pétala'].median(), color='r', linestyle='--')
eixo.annotate('mediana', xy=(df_is['comprimento_pétala'].median()+0.05, 12), fontsize=15,
             color='r')

fig.savefig('histograma_is.png', bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0,0,1,1])

df_ivg = df[df['espécie'] == 'Iris-virginica']

mu, sigma = df_ivg['comprimento_pétala'].mean(),df_ivg['comprimento_pétala'].std()

eixo.hist(df_ivg['comprimento_pétala'], bins=20)
eixo.set_title('Íris-Virginica / Histograma Comprimento da Pétala', fontsize=15)
eixo.set_xlabel('Comprimento da Pétala', fontsize=15)
eixo.grid(True)

eixo.annotate(f'$\mu = {mu:.2f}$\n$\sigma = {sigma:.2f}$',
               xy=(6.3, 7), fontsize=20)

eixo.axvline(mu, color='k', linestyle='--')
eixo.annotate('média', xy=(mu, 8), fontsize=15)

eixo.axvline(df_ivg['comprimento_pétala'].median(), color='r', linestyle='--')
eixo.annotate('mediana', xy=(df_ivg['comprimento_pétala'].median(), 7), fontsize=15,
             color='r')

fig.savefig('histograma_ivg.png', bbox_inches='tight')
```

```python
from PIL import Image

largura, altura = Image.open('histograma_ivg.png').size
combinada = Image.new("RGB",(3*largura, altura))
intervalo = 0
for imagem in map(Image.open, ['histograma_iv.png', 'histograma_is.png', 'histograma_ivg.png']):
    combinada.paste(imagem, (intervalo, 0))
    intervalo += largura
    
combinada.save('combinada.png')
```
