# Importando bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import numpy as np
```

***Criamos a Regressão Linear em todos os casos com a função do scipy, ao invés de fazer "na raça" que nem na aula**

# MRU

## Caso 11

```python
dados_vel11 = pd.read_csv('vel11.csv', sep=';', decimal=',')
dados_ac11 = pd.read_csv('ac11.csv', sep=';', decimal=',')
dados_pos11 = pd.read_csv('pos11.csv', sep=';', decimal=',')
```

## Tempo X Aceleração

```python
res = stats.linregress(x = dados_ac11['Tempo'], y = dados_ac11['Aceleração'])
x = dados_ac11['Tempo']

grafico = plt.scatter(x = x, y = dados_ac11['Aceleração'], label='dados aceleração', marker="*", color='k')

plt.plot(dados_ac11['Tempo'], res.intercept+ res.slope*x, 'purple', label='linha reg linear', linestyle='--')
plt.title('Reg Linear - Tempo x Aceleração', fontsize=22)
plt.legend()
plt.show()
```

## Tempo X Velocidade

```python
res = stats.linregress(x = dados_vel11['Tempo'], y = dados_vel11['Velocidade'])
x = dados_vel11['Tempo']

grafico = plt.scatter(x = x, y = dados_vel11.Velocidade, label='dados velocidade', marker="*", color='k')

plt.plot(dados_vel11['Tempo'], res.intercept+ res.slope*x, 'purple', label='linha reg linear', linestyle='--')
plt.title('Reg Linear - Tempo x Velocidade', fontsize=22)
plt.legend()
plt.show()
```

## Tempo X Posição

```python
res = stats.linregress(x = dados_pos11['Tempo'], y = dados_pos11['Posição'])
x = dados_pos11['Tempo']

grafico = plt.scatter(x = x, y = dados_pos11.Posição, label='dados Posição', marker="*", color='k')

plt.plot(dados_pos11['Tempo'], res.intercept+ res.slope*x, 'purple', label='linha reg linear', linestyle='--')
plt.title('Reg Linear - Tempo x Posição', fontsize=22)
plt.legend()
plt.show()
```
