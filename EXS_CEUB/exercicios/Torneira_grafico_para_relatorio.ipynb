```python
import matplotlib.pyplot as plt
import numpy as np

# criando 1000 pontos espaçados entre 0 e 81 
x = np.linspace(0, 81, 1000)

# calculando o y para cada valor x
y = -0.00429 * (x**2)

fig, ax = plt.subplots()
ax.plot(x, y, linewidth=7)
fig.set_size_inches(22, 7)
plt.title("Gráfico trajetória partícula de água", fontsize=35)
plt.xlabel(" Distância X (em cm)", fontsize=25)
plt.ylabel(" Altura Y (em cm)", fontsize=22)
```
